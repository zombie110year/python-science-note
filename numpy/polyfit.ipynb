# 拟合数据

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_known = np.array([0, 1, 2, 3, 4, 5], dtype=np.float)
y_known = np.array([2, 1, 7, 9, 0, 3], dtype=np.float)
```

```python
plt.scatter(x_known, y_known)
```

最小二乘拟合：`np.polyfit(x, y, deg, ...)`，deg 参数表示多项式的最高幂次。

```python
p = np.polyfit(x_known, y_known, 3)
```

```python
x = np.linspace(0, 10, 1000)
y = np.polyval(p, x)
```

```python
plt.plot(x, y, label="polyfig")
plt.scatter(x_known, y_known, color="r", marker="x", label="raw")
plt.legend()
```
